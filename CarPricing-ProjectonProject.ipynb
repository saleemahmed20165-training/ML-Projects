
###r2_score : "Coefficient of determination also called as R2 score is used to evaluate the performance of a linear regression model." It is the amount of the variation in the output dependent attribute which is predictable from the input independent variable(s). Best possible score is 1.0 and it can be negative (because the model can be arbitrarily worse). A constant model that always predicts the expected value of y, disregarding the input features, would get a score of 0.0.

###OneHotEncoder: Machines understand numbers, not text. We need to convert each text category to numbers in order for the machine to process them using mathematical equations.That’s primarily the reason we need to convert categorical columns to numerical columns so that a machine learning algorithm understands it. This process is called categorical encoding. One-Hot Encoding is another popular technique for treating categorical variables. It simply creates additional features based on the number of unique values in the categorical feature. Every unique value in the category will be added as a feature. One-Hot Encoding is the process of creating dummy variables.

###Pipeline : Machine Learning projects need iterative progress. For example, we clean and prepare data for modeling with transforming to the proper format, run the model, get results, improve the model/change the model and work on feature engineering, get new results, compare them with other results, etc. It is not easy and smart to do every step again and again. To solve this problem, we can use a pipeline to integrate steps of machine learning workflow. Pipelines are super useful for transforming and training data quickly.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder  # changes str to numeric
from sklearn.compose import make_column_transformer # encoded data will be transformed
from sklearn.pipeline import make_pipeline # full data is put in pipeline
```

```python
ohe=OneHotEncoder()
ohe.fit(X[["name","company","fuel_type"]])
```

```python
#ohe.categories_     # displays Y= [name, company, tuel_type]    x=Price
```

```python
col_trans=make_column_transformer((OneHotEncoder(categories=ohe.categories_),["name","company","fuel_type"]),remainder="passthrough")
# col_trans = make_column_transformer
# OneHotEncoder =  (X[["name","company","fuel_type"]])
# categories list["name","company","fuel_type"]
# remainder
```

```python
'''
**diagram‑style explanation** of how your `ColumnTransformer` works. No images needed; this is a structured, easy‑to‑visualize flow you can even paste into your notebook as documentation.
# **📊 Visual Diagram — How the ColumnTransformer Works**
                   ┌──────────────────────────────────────────────┐
                   │                Original DataFrame             │
                   │──────────────────────────────────────────────│
                   │  name   | company | fuel_type | mileage | ... │
                   └──────────────────────────────────────────────┘
                                      │  (select columns)
                                      ▼
        ┌────────────────────────────────────────────────────────────────────┐
        │                     ColumnTransformer (col_trans)                  │
        │────────────────────────────────────────────────────────────────────│
        │  1. OneHotEncoder(categories=ohe.categories_) applied to:          │
        │        → ["name", "company", "fuel_type"]                          │
        │                                                                     │
        │  2. remainder="passthrough":                                       │
        │        → all other columns (e.g., mileage, engine_size, etc.)      │
        │          are kept unchanged                                        │
        └────────────────────────────────────────────────────────────────────┘
                                      │  (transform)
                                      ▼
┌────────────────────────────────────────────────────────────────────────────────┐
│                           Transformed Feature Matrix                            │
│────────────────────────────────────────────────────────────────────────────────│
│  name_A  name_B  ...  company_X  company_Y  ...  fuel_type_P  fuel_type_Q  ... │
│────────────────────────────────────────────────────────────────────────────────│
│  mileage | engine_size | ... (unchanged numeric columns)                       │
└────────────────────────────────────────────────────────────────────────────────┘
# **🧠 What This Diagram Shows**
## **1. Input DataFrame**
You start with a mix of:
- categorical columns → `"name"`, `"company"`, `"fuel_type"`
- numeric columns → `"mileage"`, `"engine_size"`, etc.

### **2. ColumnTransformer Logic**
- The **OneHotEncoder** transforms only the three categorical columns.
- The **remainder="passthrough"** ensures all other columns flow through untouched.
### **3. Output Matrix**
After transformation, you get:
- Many new **binary one‑hot columns**
  (e.g., `company_Toyota`, `company_Honda`, etc.)
- All numeric columns preserved exactly as they were.

This final matrix is what you feed into your model.

If you'd like, I can also create:
- a **pipeline diagram** showing how this fits into training
- a **Colab‑ready code block** that prints the transformed feature names
- a **side‑by‑side before/after table** to make it even clearer
'''
```

```python
lr=LinearRegression()
```

```python
pipe=make_pipeline(col_trans,lr)  # create pipe using col_trans, using linear regression
# col_trans is make_column_transformer using (ohe using (ohe.categories [name, company, fuel_type])) remainder=passthrough
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred=pipe.predict(X_test)  # calculate the score of prediction using test data
```

```python
r2_score(y_test,y_pred)  # prediction is correct for our test
```

```python
pd.DataFrame({'Actual Price':y_test,'Predicted Price':y_pred})
```
